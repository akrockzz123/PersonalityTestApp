```python
text_path = "/content/drive/MyDrive/speech.txt"

def load_text(filename):
  with open(filename, mode='r', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None) as f:
    return f.read().lower()
```

```python
data = load_text(text_path)

print(data)
```

```python
def generateTable(data, k= 4):
   T = {}
   for i in range(len(data)-k):
     x = data[i:i+k]
     y = data[i+k]
    ## printing x and y print(x," ",y)
     
     if T.get(x) is None:
       T[x] = {}
       T[x][y] = 1
     else:
       if T[x].get(y) is None:
         T[x][y] = 1
       else: 
         T[x][y] += 1

   return T
```

```python
T = generateTable(data)
```

```python
##Sampling
import numpy as np

def sample_next(ctx,T,k):
  cts = ctx[-k,:]
  if T.get(cts) is None :
    return " "
  else:
    posssible_chars = list(T[cts].keys())
    possible_values = list(T[cts].values())
  
  return np.random_choice(possibel_chars,p = possible_values)
  
```

```python
def convertfreq2prob(T):
  for kx in T.keys():
     s = float(sum(T[kx].values()))
     for x in T[kx].keys():
       T[kx][x] = T[kx][x]/s;
  
  return T;
```

```python
convertfreq2prob(T)
```

```python
def trainMarkovChain(text,k=4):
  T = generateTable(text,k)
  T = convertfreq2prob(T)
  return T
```
